```python
import pandas as pd
from binance.client import Client
import pickle
```

```python
client = Client()

def get_klines(tipo,start,end,symbol,period):
    df = client.get_historical_klines(symbol=symbol, interval=period, start_str=start, end_str=end)
    df = pd.DataFrame(df)
    df = df.iloc[:, :6]
    df.columns = ["datetime", "open", "high", "low", "close", "volume"]
    df['open'] = df['open'].astype('float')
    df['high'] = df['high'].astype('float')
    df['low'] = df['low'].astype('float')
    df['close'] = df['close'].astype('float')
    df['volume'] = df['volume'].astype('float')
    df['datetime'] = pd.to_datetime(df['datetime'], unit='ms') - pd.Timedelta('3 hr')
    df=df.set_index('datetime')
    return df
```

```python
folder = '../bt_sources/p0m01/'
period = '1m'

periods = [{'tipo':'Lateral',
            'start':'2023-06-20',
            'end':'2023-10-09',
            'symbols': {
                    'DOGEUSDT',
                    'SOLUSDT', 
                    'BNBUSDT', 
                    'ETHUSDT', 
                    'ADAUSDT', 
                    'TRXUSDT', 

                }
            },
            {'tipo':'Alcista',
            'start':'2020-09-07',
            'end':'2021-05-03',
            'symbols': {
                        'BTCUSDT', 
                        'ETHUSDT', 
                        'BNBUSDT', 
                        'XRPUSDT', 
                        'SOLUSDT', 
                        'ADAUSDT', 
                        'AVAXUSDT', 
                        'TRXUSDT', 
                        'DOTUSDT', 
                        'DOGEUSDT', 
                }
            },
            {'tipo':'Bajista',
            'start':'2021-04-12',
            'end':'2021-07-12',
            'symbols': {
                        'BTCUSDT', 
                        'ETHUSDT', 
                        'BNBUSDT', 
                        'ADAUSDT', 
                        'TRXUSDT', 
                        'DOTUSDT', 
                        'XRPUSDT', 

                }
            },
            ]

"""periods = [{'tipo':'Lateral',
            'start':'2023-06-20',
            'end':'2023-06-21',
            'symbols': {
                    'DOGEUSDT',
                }
            },
            {'tipo':'Alcista',
            'start':'2020-09-07',
            'end':'2020-09-08',
            'symbols': {
                        'BTCUSDT', 
                        'ETHUSDT',
                }
            },
            ] """
```

```python
for p in periods:
    tipo = p['tipo']
    start = p['start']
    end = p['end']
    utc_start = start+' 00:00:00-03:00'
    utc_end = end+' 23:59:59-03:00'
    for symbol in p['symbols']:
        file_dump = f'{folder}{tipo}_{symbol}_{period}_{start}_{end}.DataFrame'
        print(file_dump,end=' -> ')
        df = get_klines(tipo,utc_start,utc_end,symbol,period)
        print('OK',end=' -> ')
        with open(file_dump, 'wb') as f:
            print('File',end=' -> ')
            pickle.dump(df, f)
            print('Ready')
```
